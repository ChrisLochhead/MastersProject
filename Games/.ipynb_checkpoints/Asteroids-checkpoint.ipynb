```python
import sys
import pygame
import pygame_menu
import random
import math
import numpy as np
import os
#Change path to import modules
sys.path.append(r"C:/Users/chris/OneDrive/Documents/Masters Project/")
from ipynb.fs.full.Dependencies.Training_Game import Training_Game
#Change directory
os.chdir(r"C:/Users/chris/OneDrive/Documents/Masters Project/")
pygame.init()
```

```python
class Asteroids(Training_Game):
    
    def split_asteroid(self, a):
        # Split asteroid
        if a.t > 10:
            self.asteroids.append(Asteroid(a.x, a.y, a.t-10, self.width, self.height))
            self.asteroids.append(Asteroid(a.x, a.y, a.t-10, self.width, self.height))
            self.score += 20 if a.t == 30 else 50
        else:
            self.score +=100
        self.asteroids.remove(a)
        
    def destroy_player(self):
        self.player_pieces.append(deadPlayer(self.player.x, self.player.y, 5 * 10 / (2 * math.cos(math.atan(1 / 3)))))
        self.player_pieces.append(deadPlayer(self.player.x, self.player.y, 5 * 10 / (2 * math.cos(math.atan(1 / 3)))))
        self.player_pieces.append(deadPlayer(self.player.x, self.player.y, 10))

        # Kill player
        self.player.player_state = "Died"
        self.gameState = "Game Over"
    
    def get_state(self):
        self.pixel_array = pygame.PixelArray(self.surface)
        #print("surface information before: \ndimensions: ", self.pixel_array.ndim, "\nShape: ",
        #    self.pixel_array.shape, "\n")
        self.pixel_array.close()
        #print("surface information after: \ndimensions: ", self.pixel_array.ndim, "\nShape: ",
        #    self.pixel_array.shape, "\n")
        return self.pixel_array.flatten()
    
    def execute_action(self, action):
        if action == 0:
            self.turn_left()
        if action == 1:
            self.turn_right()
        if action == 2:
            self.shoot()
        if action == 3:
            self.accelerate()
        if action == 4:
            self.decelerate()
        if action == 5:
            self.stop_rotating()
            
    def accelerate(self):
        self.player.thrust = True
    def decelerate(self):
        self.player.thrust = False
    def shoot(self):
        self.bullets.append(Bullet(self.player.x, self.player.y, self.player.dir,
                                    self.width, self.height))
    def turn_left(self):
        self.player.rtspd = -player_max_rtspd
    def turn_right(self):
        self.player.rtspd = player_max_rtspd
    def stop_rotating(self):
        self.player.rtspd = 0
        
    def update(self):
        # User inputs
        if self.training_mode == False:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    self.gameState = "Exit"
                if event.type == pygame.KEYDOWN:
                    if self.player.player_state != "Died":
                        if event.key == pygame.K_UP:
                            self.accelerate()
                        if event.key == pygame.K_LEFT:
                            self.turn_left()
                        if event.key == pygame.K_RIGHT:
                            self.turn_right()
                        if event.key == pygame.K_SPACE and len(self.bullets) < self.bullet_capacity:
                            if event.key == pygame.K_SPACE and len(self.bullets) < self.bullet_capacity:
                                self.shoot()
                    if self.gameState == "Game Over":
                        if event.key == pygame.K_r:
                            #GO back to menu here
                            self.gameState = "Exit"
                        if event.key == pygame.K_a:
                            self.reset()
                if self.player.player_state != "Died":
                    if event.type == pygame.KEYUP:
                        if event.key == pygame.K_UP:
                            self.decelerate()
                        if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                            self.stop_rotating()

        # Update player
        self.player.updatePlayer()

        for a in self.asteroids:
            a.update()
            if self.player.player_state != "Died":
                if isColliding(self.player.x, self.player.y, a.x, a.y, a.size):
                    self.destroy_player()
                    self.gameState = "Game Over"
                    self.split_asteroid(a)

        # Update ship fragments
        for f in self.player_pieces:
            f.update()
            if f.x > self.width or f.x < 0 or f.y > self.height or f.y < 0:
                self.player_pieces.remove(f)

        # Check for end of stage
        if len(self.asteroids) == 0 and self.saucer.state == "Dead":
            if self.next_level_delay < 30:
                self.next_level_delay += 1
            else:
                self.stage += 1
                self.intensity = 0
                # Spawn asteroid away of center
                for i in range(self.stage):
                    xTo = self.width / 2
                    yTo = self.height / 2
                    while xTo - self.width / 2 < self.width / 4 and yTo \
                    - self.height / 2 < self.height / 4:
                        xTo = random.randrange(0, self.width)
                        yTo = random.randrange(0, self.height)
                    self.asteroids.append(Asteroid(xTo, yTo, 30, self.width, self.height))
                self.next_level_delay = 0

        # Update intensity
        if self.intensity < self.stage * 450:
            self.intensity += 1

        # Saucer
        if self.saucer.state == "Dead":
            if random.randint(0, 6000) <= (self.intensity * 2) / (self.stage * 9) and self.next_level_delay == 0:
                self.saucer.createSaucer()
                # Only small saucers >40000
                if self.score >= 40000:
                    self.saucer.type = "Small"
        else:
            # Set saucer targer dir
            acc = small_saucer_accuracy * 4 / self.stage
            self.saucer.bdir = math.degrees(math.atan2(-self.saucer.y + self.player.y,
                                                       -self.saucer.x + self.player.x) + math.radians(random.uniform(acc, -acc)))

            self.saucer.updateSaucer()
            self.saucer.drawSaucer(self.gameDisplay)

            # Check for collision w/ asteroid
            for a in self.asteroids:
                if isColliding(self.saucer.x, self.saucer.y, a.x, a.y, a.size + self.saucer.size):
                    # Set saucer state
                    self.saucer.state = "Dead"
                    # Split asteroid 
                    self.split_asteroid(a)

            # Check for collision w/ bullet
            for b in self.bullets:
                if isColliding(b.x, b.y, self.saucer.x, self.saucer.y, self.saucer.size):
                    # Add points
                    if self.saucer.type == "Large":
                        self.score += 200
                    else:
                        self.score += 1000
                    # Set saucer state
                    self.saucer.state = "Dead"
                    # Remove bullet
                    self.bullets.remove(b)

            # Check collision w/ player
            if isColliding(self.saucer.x, self.saucer.y, self.player.x, self.player.y, self.saucer.size):
                if self.player.player_state != "Died":
                    self.destroy_player()

            # Saucer's bullets
            for b in self.saucer.bullets:
                # Update bullets
                b.update()

                # Check for collision w/ asteroids
                for a in self.asteroids:
                    if isColliding(b.x, b.y, a.x, a.y, a.size):
                        self.split_asteroid(a)
                        self.saucer.bullets.remove(b)
                        break

                # Check for collision w/ player
                if isColliding(self.player.x, self.player.y, b.x, b.y, 5):
                    if self.player.player_state != "Died":
                        self.destroy_player()
                        # Remove bullet
                        self.saucer.bullets.remove(b)
                        
                if b.life <= 0:
                    try:
                        self.saucer.bullets.remove(b)
                    except ValueError:
                        continue

        # Bullets
        for b in self.bullets:
            # Update bullets
            b.update()

            # Check for bullets collide w/ asteroid
            for a in self.asteroids:
                if b.x > a.x - a.size and b.x < a.x + a.size and b.y > a.y - a.size and b.y < a.y + a.size:
                    # Split asteroid
                    self.split_asteroid(a)
                    self.bullets.remove(b)
                    break
                    
            if b.life <= 0:
                try:
                    self.bullets.remove(b)
                except ValueError:
                    continue

    def render(self):
        # Reset display
        self.gameDisplay.fill(black)
        
        #Draw the asteroids
        for a in self.asteroids:
            a.render(self.gameDisplay)
        
        # Update ship fragments
        for f in self.player_pieces:
            f.render(self.gameDisplay)
            
        # Bullets
        for b in self.bullets:
            # Update bullets
            b.render(self.gameDisplay)
        
        #Saucer bullets
        for b in self.saucer.bullets:
            b.render(self.gameDisplay)
            
        # Draw player
        if self.gameState != "Game Over":
            if not self.player.player_state == "Died":
                self.player.drawPlayer(self.gameDisplay)
        else:
            self.drawText("Game Over", white, self.width / 2, self.height / 2, 100)
            self.drawText("Press \"R\" to Return to the menu.", white, self.width / 2, self.height / 2 + 100, 50)

        # Draw score
        self.drawText(str(self.score), white, 60, 20, 40, False)
        # Update screen
        pygame.display.update()

        # Tick fps
        self.timer.tick(30)
    
    def reset(self):
        del self.player
        self.player = Asteroid_Player(self.width / 2, self.height / 2, self.width, self.height)
        self.next_level_delay = 0
        self.bullets.clear()
        self.asteroids = []
        self.stage = 3
        self.score = 0
        self.intensity = 0
        self.gameState = "Playing"
        self.player_pieces = []
        
    def __init__(self, width, height, screen, surface, mode):
        self.gameDisplay = pygame.display.set_mode((width, height))
        self.timer = pygame.time.Clock()
        # Init variables
        self.gameState = "Playing"
        self.player_pieces = []
        self.width = width
        self.height = height
        self.next_level_delay = 0
        self.bullet_capacity = 4
        self.bullets = []
        self.asteroids = []
        self.stage = 3
        self.score = 0
        self.intensity = 0
        self.training_mode = mode
        player = Asteroid_Player(width / 2, height / 2, width, height)
        self.player = player
        self.saucer = Saucer(width, height)

        # Main loop
        if self.training_mode == False:
            while self.gameState != "Exit":
                self.update()
                self.render()
           
    # Create function to draw texts
    def drawText(self, msg, color, x, y, s, center=True):
        screen_text = pygame.font.SysFont("Calibri", s).render(msg, True, color)
        if center:
            rect = screen_text.get_rect()
            rect.center = (x, y)
        else:
            rect = (x, y)
        self.gameDisplay.blit(screen_text, rect)
```

```python
#Auxillary classes for the Asteroids game mode
# Initialize constants
white = (255, 255, 255)
red = (255, 0, 0)
black = (0, 0, 0)

#player_size = 10
fd_fric = 0.5
bd_fric = 0.1
player_max_speed = 20
player_max_rtspd = 10
bullet_speed = 15
saucer_speed = 5
small_saucer_accuracy = 10

# Create funtion to chek for collision
def isColliding(x, y, xTo, yTo, size):
    if x > xTo - size and x < xTo + size and y > yTo - size and y < yTo + size:
        return True
    return False

def wrap_object(x, y, width, height):
    # Check for wrapping
    if x > width:
        x = 0
    elif x < 0:
        x = width
    elif y > height:
        y = 0
    elif y < 0:
        y = height
    return x, y

# Create class asteroid
class Asteroid:
    def __init__(self, x, y, t, width, height):
        self.x = x
        self.y = y
        self.size = t
        self.t = t
        self.width = width
        self.height = height
        
        # Make random speed and direction
        self.speed = random.uniform(1, (40 - self.size) * 4 / 15)
        self.dir = random.randrange(0, 360) * math.pi / 180

        # Make random asteroid sprites
        full_circle = random.uniform(18, 36)
        dist = random.uniform(self.size / 2, self.size)
        self.vertices = []
        while full_circle < 360:
            self.vertices.append([dist, full_circle])
            dist = random.uniform(self.size / 2, self.size)
            full_circle += random.uniform(18, 36)
    
    def render(self, gameDisplay):
        # Draw asteroid
        for v in range(len(self.vertices)):
            if v == len(self.vertices) - 1:
                next_v = self.vertices[0]
            else:
                next_v = self.vertices[v + 1]
            this_v = self.vertices[v]
            pygame.draw.line(gameDisplay, white, (self.x + this_v[0] * math.cos(this_v[1] * math.pi / 180),
                                                  self.y + this_v[0] * math.sin(this_v[1] * math.pi / 180)),
                             (self.x + next_v[0] * math.cos(next_v[1] * math.pi / 180),
                              self.y + next_v[0] * math.sin(next_v[1] * math.pi / 180)))
            
    def update(self):
        # Move asteroid
        self.x += self.speed * math.cos(self.dir)
        self.y += self.speed * math.sin(self.dir)
        self.x, self.y = wrap_object(self.x, self.y, self.width, self.height)

# Create class bullet
class Bullet:
    def __init__(self, x, y, direction, width, height):
        self.x = x
        self.y = y
        self.dir = direction
        self.life = 30
        self.width = width
        self.height = height
    
    def render(self, gameDisplay):
        pygame.draw.circle(gameDisplay, white, (int(self.x), int(self.y)), 3)
        
    def update(self):
        # Moving
        self.x += bullet_speed * math.cos(self.dir * math.pi / 180)
        self.y += bullet_speed * math.sin(self.dir * math.pi / 180)
        self.x, self.y = wrap_object(self.x, self.y, self.width, self.height)
        self.life -= 1


# Create class saucer
class Saucer:
    def __init__(self, width, height):
        self.x = 0
        self.y = 0
        self.state = "Dead"
        self.type = "Large"
        self.dirchoice = ()
        self.bullets = []
        self.cd = 0
        self.bdir = 0
        self.soundDelay = 0
        self.width = width
        self.height = height
        
    def updateSaucer(self):
        # Move player
        self.x += saucer_speed * math.cos(self.dir * math.pi / 180)
        self.y += saucer_speed * math.sin(self.dir * math.pi / 180)

        # Choose random direction
        if random.randrange(0, 100) == 1:
            self.dir = random.choice(self.dirchoice)

        # Wrapping
        if self.y < 0:
            self.y = self.height
        elif self.y > self.height:
            self.y = 0
        if self.x < 0 or self.x > self.width:
            self.state = "Dead"

        # Shooting
        if self.type == "Large":
            self.bdir = random.randint(0, 360)
        if self.cd == 0:
            self.bullets.append(Bullet(self.x, self.y, self.bdir, self.width, self.height))
            self.cd = 30
        else:
            self.cd -= 1

    def createSaucer(self):
        # Create saucer
        # Set state
        self.state = "Alive"

        # Set random position
        self.x = random.choice((0, self.width))
        self.y = random.randint(0, self.height)

        # Set random type
        if random.randint(0, 1) == 0:
            self.type = "Large"
            self.size = 20
        else:
            self.type = "Small"
            self.size = 10

        # Create random direction
        if self.x == 0:
            self.dir = 0
            self.dirchoice = (0, 45, -45)
        else:
            self.dir = 180
            self.dirchoice = (180, 135, -135)

        # Reset bullet cooldown
        self.cd = 0

    def drawSaucer(self, gameDisplay):
        # Draw saucer
        pygame.draw.polygon(gameDisplay, white,
                            ((self.x + self.size, self.y),
                             (self.x + self.size / 2, self.y + self.size / 3),
                             (self.x - self.size / 2, self.y + self.size / 3),
                             (self.x - self.size, self.y),
                             (self.x - self.size / 2, self.y - self.size / 3),
                             (self.x + self.size / 2, self.y - self.size / 3)), 1)
        pygame.draw.line(gameDisplay, white,
                         (self.x - self.size, self.y),
                         (self.x + self.size, self.y))
        pygame.draw.polygon(gameDisplay, white,
                            ((self.x - self.size / 2, self.y - self.size / 3),
                             (self.x - self.size / 3, self.y - 2 * self.size / 3),
                             (self.x + self.size / 3, self.y - 2 * self.size / 3),
                             (self.x + self.size / 2, self.y - self.size / 3)), 1)


# Create class for shattered ship
class deadPlayer:
    def __init__(self, x, y, l):
        self.angle = random.randrange(0, 360) * math.pi / 180
        self.dir = random.randrange(0, 360) * math.pi / 180
        self.rtspd = random.uniform(-0.25, 0.25)
        self.x = x
        self.y = y
        self.lenght = l
        self.speed = random.randint(2, 8)
    
    def render(self, gameDisplay):
        pygame.draw.line(gameDisplay, white,
                 (self.x + self.lenght * math.cos(self.angle) / 2,
                  self.y + self.lenght * math.sin(self.angle) / 2),
                 (self.x - self.lenght * math.cos(self.angle) / 2,
                  self.y - self.lenght * math.sin(self.angle) / 2))
        
    def update(self):
        self.angle += self.rtspd
        self.x += self.speed * math.cos(self.dir)
        self.y += self.speed * math.sin(self.dir)


# Class player
class Asteroid_Player:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.hspeed = 0
        self.vspeed = 0
        self.dir = -90
        self.rtspd = 0
        self.thrust = False
        self.player_state = "Alive"
        self.display_width = width
        self.display_height = height

    def updatePlayer(self):
        # Move player
        speed = math.sqrt(self.hspeed**2 + self.vspeed**2)
        if self.thrust:
            if speed + fd_fric < player_max_speed:
                self.hspeed += fd_fric * math.cos(self.dir * math.pi / 180)
                self.vspeed += fd_fric * math.sin(self.dir * math.pi / 180)
            else:
                self.hspeed = player_max_speed * math.cos(self.dir * math.pi / 180)
                self.vspeed = player_max_speed * math.sin(self.dir * math.pi / 180)
        else:
            if speed - bd_fric > 0:
                change_in_hspeed = (bd_fric * math.cos(self.vspeed / self.hspeed))
                change_in_vspeed = (bd_fric * math.sin(self.vspeed / self.hspeed))
                if self.hspeed != 0:
                    if change_in_hspeed / abs(change_in_hspeed) == self.hspeed / abs(self.hspeed):
                        self.hspeed -= change_in_hspeed
                    else:
                        self.hspeed += change_in_hspeed
                if self.vspeed != 0:
                    if change_in_vspeed / abs(change_in_vspeed) == self.vspeed / abs(self.vspeed):
                        self.vspeed -= change_in_vspeed
                    else:
                        self.vspeed += change_in_vspeed
            else:
                self.hspeed = 0
                self.vspeed = 0
        self.x += self.hspeed
        self.y += self.vspeed

        # Check for wrapping
        self.x, self.y = wrap_object(self.x, self.y, self.display_width, self.display_height)
        # Rotate player
        self.dir += self.rtspd

    def drawPlayer(self, gameDisplay):
        a = math.radians(self.dir)
        x = self.x
        y = self.y
        s = 10
        t = self.thrust
        # Draw player
        for i in range(0,2):
            pygame.draw.line(gameDisplay, white,
                (x - (s * math.sqrt(130) / 12) * math.cos(math.atan(7 / 9) + a),
                y - (s * math.sqrt(130) / 12) * math.sin(math.atan(7 / 9) + a)),
                (x + s * math.cos(a), y + s * math.sin(a)))
        
        pygame.draw.line(gameDisplay, white,
                         (x - (s * math.sqrt(130) / 12) * math.cos(math.atan(7 / 9) - a),
                          y + (s * math.sqrt(130) / 12) * math.sin(math.atan(7 / 9) - a)),
                         (x + s * math.cos(a), y + s * math.sin(a)))

        pygame.draw.line(gameDisplay, white,
                         (x - (s * math.sqrt(2) / 2) * math.cos(a + math.pi / 4),
                          y - (s * math.sqrt(2) / 2) * math.sin(a + math.pi / 4)),
                         (x - (s * math.sqrt(2) / 2) * math.cos(-a + math.pi / 4),
                          y + (s * math.sqrt(2) / 2) * math.sin(-a + math.pi / 4)))
        if t:
            pygame.draw.line(gameDisplay, white,
                             (x - s * math.cos(a),
                              y - s * math.sin(a)),
                             (x - (s * math.sqrt(5) / 4) * math.cos(a + math.pi / 6),
                              y - (s * math.sqrt(5) / 4) * math.sin(a + math.pi / 6)))
            pygame.draw.line(gameDisplay, white,
                             (x - s * math.cos(-a),
                              y + s * math.sin(-a)),
                             (x - (s * math.sqrt(5) / 4) * math.cos(-a + math.pi / 6),
                              y + (s * math.sqrt(5) / 4) * math.sin(-a + math.pi / 6)))
```
